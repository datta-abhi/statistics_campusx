### Q-1: What is the probability distribution of age in the insurance dataset?

-- Find out using the KDE method

```python
plt.hist(df['age'],bins = 10,density = True)
plt.show()
```

```python
from sklearn.neighbors import KernelDensity
```

```python
# Fitting data to a KDE
model = KernelDensity(bandwidth=2)
model.fit(df['age'].dropna().to_numpy().reshape(-1,1))

# creating x and y data for plotting
x = np.linspace(df['age'].min(),df['age'].max(),500)
y_log_kde = model.score_samples(x.reshape(-1,1))
y_kde = np.exp(y_log_kde)

# Plotting kde
plt.plot(x,y_kde)
plt.hist(df['age'],bins = 10,density = True)
plt.show()
```

### Q-3: Plot distribution plot of claim for Smoker and non smoker. What changes you see in the plot?

```python
# Create a figure
plt.figure(figsize=(12, 5))  # Adjust the figure size as needed

# Subplot 1 - Histogram with hue
plt.subplot(1, 2, 1)
sns.histplot(data=df, x='claim', hue='smoker')

# Subplot 2 - KDE Plot with hue
plt.subplot(1, 2, 2)
sns.kdeplot(data=df, x='claim', hue='smoker')

# Show the plot
plt.show()
```

### Q-4: Plot the 2D probability density plot of claim and age in the insurance dataset?

```python
# code here
plt.figure(figsize=(6,4))
sns.kdeplot(data = df, x = 'age', y = 'claim',cmap = 'icefire',fill = True)
plt.show()
```
