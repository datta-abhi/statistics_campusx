```python
import numpy as np
import pandas as pd
```

# PMF, CDF and PDF

## Simulating 1 die roll

```python
size = 99999
a1 = np.random.randint(1,7,size = size)

s1 = pd.Series(a1)
s2 = np.round(s1.value_counts().sort_index()/len(s1),4)
s2.plot(kind = 'bar')
```

## CDF of 1 die roll

```python
s3 = np.cumsum(s2)
s3.plot(kind = 'bar')
```

## Simulating 2 die rolls

```python
size = 99999
a1 = np.random.randint(1,7,size = size)
a2 = np.random.randint(1,7,size = size)
a3 = (a1+a2)
s1 = pd.Series(a3)
s2 = np.round(s1.value_counts().sort_index()/len(s1),4)
s2.plot(kind = 'bar')
```

```python
s3 = np.cumsum(s2)
s3.plot(kind = 'bar')
```
