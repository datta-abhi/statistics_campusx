# PMF, CDF and PDF

## Simulating 1 die roll

```python
size = 99999
a1 = np.random.randint(1,7,size = size)

s1 = pd.Series(a1)
s2 = np.round(s1.value_counts().sort_index()/len(s1),4)
s2.plot(kind = 'bar')
```

## CDF of 1 die roll

```python
s3 = np.cumsum(s2)
s3.plot(kind = 'bar')
```

## Simulating 2 die rolls

```python
size = 99999
a1 = np.random.randint(1,7,size = size)
a2 = np.random.randint(1,7,size = size)
a3 = (a1+a2)
s1 = pd.Series(a3)
s2 = np.round(s1.value_counts().sort_index()/len(s1),4)
s2.plot(kind = 'bar')
```

```python
s3 = np.cumsum(s2)
s3.plot(kind = 'bar')
```

## Parametric Density Estimation

```python
size = 999
pop_mean = 50
pop_sd = 20
# Create a sample of datapoints whose population mean is 50 and population sd is 5
sample = np.random.normal(loc = pop_mean,scale = pop_sd,size = size)
print('Population mean: {}, population sd: {}'.format(pop_mean,pop_sd))
print('Sample mean: {}, sample sd: {}'.format(round(sample.mean(),2),round(sample.std(),2)))
```

```python
sample
```

```python
# histogram of sample
plt.hist(sample,bins = 10)
```

## Create a normal distribution pdf curve with sample mean and std dev

```python
sample_mean = sample.mean()
sample_std = sample.std()
```

```python
from scipy.stats import norm
```

```python
dist = norm(sample_mean,sample_std)
```

```python
# Plotting the pdf based on this normal distribution
x = np.linspace(sample.min(),sample.max(),1000)
y = dist.pdf(x)
```

```python
plt.plot(x,y)
plt.hist(sample,bins = 10,density=True)
```
