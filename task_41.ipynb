````python
from sklearn.datasets import load_wine
data = load_wine()
wine = pd.DataFrame(data = data.data,columns=data.feature_names)
````

````python
wine.head(2)
````

````python
data.keys()
````

````python
wine['target'] = data.target
````

````python
wine.info()
````

````python
wine['target'].value_counts(normalize=True)
````

````python
cols = [i for i in wine.columns if i != 'target']
print(cols)
````

````python
for col in cols:
    plt.figure(figsize=(5,3))
    print(sns.kdeplot(data = wine,x = col,hue = 'target'))
    print('*'*30)
    plt.show()
````

### Based on distribution of these features, flavanoids is the least overlapping and most distinctive with respect to target.

````python
# Proline
# Flavanoids
# Color intensity
# OD280/OD315 of diluted wines
# Alcohol
````

````python
sns.jointplot(data = wine, x = 'proline',y = 'alcohol', hue = 'target',kind = 'kde')
````

````python
cols_jointplot = ['proline', 'flavanoids','color_intensity','od280/od315_of_diluted_wines','alcohol']

for main_col in cols_jointplot:
    
    plt.figure(figsize=(16, 4))  # Adjust the figure size outside the loop
    k = 1
    for col in cols_jointplot:
        if col != main_col:
            plt.subplot(1, 4, k)
            sns.kdeplot(data=wine, x=main_col, y=col, hue='target')
            
            k += 1  
    plt.tight_layout()  # Adjust subplot parameters to give specified padding
    plt.show()  # Show the plot after each main_col iteration 
    print('_'*100)
````

### Based on 2d kde plots,   'proline' - 'od280/od315_of_diluted_wines' make the best feature pair for distinction between 3 wine classes
